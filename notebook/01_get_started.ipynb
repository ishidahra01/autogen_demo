# AutoGen v0.2 sample

## ConversableAgent
Support the followning components:
1. A list of LLMs
1. A code executor
1. A function and tool executor
1. A component for keeping human-in-the-loop

### LLMs
enable agents to converse in natural languages and transform between structured and unstructured text

```python
import os

from autogen import ConversableAgent

config_list = [
  {
    "model": os.getenv("AZURE_OPENAI_CHAT_DEPLOYMENT_NAME"),
    "api_type": "azure",
    "api_key": os.environ['AZURE_OPENAI_API_KEY'],
    "base_url": os.environ['AZURE_OPENAI_ENDPOINT'],
    "api_version": os.environ['AZURE_OPENAI_API_VERSION']
  }
]

agent = ConversableAgent(
    "chatbot",
    llm_config={"config_list": config_list},
    code_execution_config=False,  # Turn off code execution, by default it is off.
    function_map=None,  # No registered functions, by default it is None.
    human_input_mode="NEVER",  # Never ask for human input.
)
```

```python
reply = agent.generate_reply(messages=[{"content": "Tell me a joke.", "role": "user"}])
print(reply)
```

## Terminating Conversations Between Agents
In this chapter, we will explore how to terminate a conversation between AutoGen agents.

Currently there are two broad mechanism to control the termination of conversations between agents:

1. Specify parameters in initiate_chat: When initiating a chat, you can define parameters that determine when the conversation should end.
1. Configure an agent to trigger termination: When defining individual agents, you can specify parameters that allow agents to terminate of a conversation based on particular (configurable) conditions.

### Parameters in `initiate_chat`

```python
import os

from autogen import ConversableAgent
```

```python
hiro = ConversableAgent(
    "hiro",
    system_message="あなたはAIの専門家で、自社のAIソリューション開発を技術面でリードする役割です。技術戦略の意思決定をして、強固なソリューションを計画、実装する責任を持ちます。",
    llm_config={"config_list": config_list},
    human_input_mode="NEVER",  # Never ask for human input.
)

aki = ConversableAgent(
    "aki",
    system_message="あなたはビジネスの専門家で、自社のAIソリューション開発をビジネス戦略立案、プロジェクトマネジメント、マーケティングをリードする意思決定者です。技術面の提案をビジネス、ROIの観点から評価、調整してビジネスを成功に導く責任を持ちます。",
    llm_config={"config_list": config_list},
    human_input_mode="NEVER",  # Never ask for human input.
)
```

## Allowing Human Feedback in Agents
In AutoGen’s ConversableAgent, the human-in-the-loop component sits in front of the auto-reply components. It can intercept the incoming messages and decide whether to pass them to the auto-reply components or to provide human feedback. The figure below illustrates the design.

### Human Input Modes
Currently AutoGen supports three modes for human input. The mode is specified through the human_input_mode argument of the ConversableAgent. The three modes are:

1. NEVER: human input is never requested.
1. TERMINATE (default): human input is only requested when a termination condition is met. Note that in this mode if the human chooses to intercept and reply, the conversation continues and the counter used by max_consecutive_auto_reply is reset.
1. ALWAYS: human input is always requested and the human can choose to skip and trigger an auto-reply, intercept and provide feedback, or terminate the conversation. Note that in this mode termination based on max_consecutive_auto_reply is ignored.

#### Human Input Mode = `NEVER`

```python
hiro = ConversableAgent(
    "hiro",
    system_message="あなたはAIの専門家で、自社のAIソリューション開発を技術面でリードする役割です。技術戦略の意思決定をして、強固なソリューションを計画、実装する責任を持ちます。",
    llm_config={"config_list": config_list},
    human_input_mode="NEVER",  # Never ask for human input.
)

aki = ConversableAgent(
    "aki",
    system_message="あなたはビジネスの専門家で、自社のAIソリューション開発をビジネス戦略立案、プロジェクトマネジメント、マーケティングをリードする意思決定者です。技術面の提案をビジネス、ROIの観点から評価、調整してビジネスを成功に導く責任を持ちます。",
    llm_config={"config_list": config_list},
    human_input_mode="NEVER",  # Never ask for human input.
    max_consecutive_auto_reply=1,  # Limit the number of consecutive auto-replies to 1.
)
```

```python
aki.initiate_chat(hiro, message="生成AIを使って、新しいAIソリューションを開発するプロジェクトを立ち上げたいと考えています。アーキテクチャを提案してください。")
```

#### Human Input Mode = `ALWAYS`

```python
human_proxy = ConversableAgent(
    "human_proxy",
    llm_config=False,  # no LLM used for human proxy
    human_input_mode="ALWAYS",  # always ask for human input
)

# Start a chat with the agent
result = human_proxy.initiate_chat(
    hiro, 
    message="コンシューマ向けのAI製品を開発するプロジェクトを立ち上げたいと考えています。どのような製品を開発するべきか、技術的な観点から提案してください。",
)
```
